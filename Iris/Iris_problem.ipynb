This is the first approach in machine learning, using the Iris dataset to create the first machine learning problem in 
the book Introduction to Machine Learning with Python.

```python
#Loading the libraries to use:
import pandas as pd
import numpy as np
from scipy import sparse
import  matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
```

First we are going to read the data:

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

For the analysis we are going the split the information into a train and test set, using the 75% to train the model

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = \
    train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

With the information splitted as train and test set, now we want to understand our information. For this we are going to 
create some plots 
#### Creating a dataset with the information
Before creating the plots, we need to create a dataframe so we can easily create the visualizations

```python
df = sns.load_dataset("iris")
sns.pairplot(df, hue="species")
```

Now that we know hoy our data, we are going to try to create a machine learning model using de K-nearest Neighbor

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
```

#### Using the model
To try the model we need to make a prediction so we are going to create an object with some random information to test it

```python
X_new = np.array([[5,2.9,1,0.2]])
prediction = knn.predict(X_new)
print("Predicted target name: {}".format(iris_dataset['target_names'][prediction]))
```

Now we are going to test the accuracy of the model to see if the prediction can be trusted

```python
y_pred = knn.predict(X_test)
print("The accuracy of the model is :{:.2f}".format(knn.score(X_test, y_test)))
```
